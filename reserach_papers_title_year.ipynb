```python
import requests
from bs4 import BeautifulSoup
import csv
import re

url = 'https://www.ashoka.edu.in/research-listing/'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

papers = []

# Scrape the research paper details
for item in soup.find_all('div', class_='research-list'):
    title_tag = item.find('h3', class_='research-lists-heading')
    title = title_tag.get_text(strip=True)

    date_tag = item.find('li', text=re.compile(r'\w+ \d{1,2}, \d{4}'))
    if date_tag:
        year = date_tag.get_text(strip=True)
    else:
        year = 'N/A'
    print("Date Tag:", date_tag) # Debugging: Print date_tag

    authors_tag = date_tag.find_next('li') if date_tag else None
    if authors_tag:
        authors = authors_tag.get_text(strip=True)
    else:
        authors = 'N/A'
    print("Authors Tag:", authors_tag) # Debugging: Print authors_tag

    papers.append([title, year, authors])

# Write data to CSV
csv_filename = 'ashoka_research_papers.csv'
with open(csv_filename, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Title', 'Year', 'Authors'])
    writer.writerows(papers)

print(f'Data saved to {csv_filename}')
```
